```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import os
import pandas as pd 
from io import StringIO
import numpy as np
from bs4 import BeautifulSoup
```

```python
chromedriver_path = "chromedriver.exe"
```

```python
# Inicializar el servicio de ChromeDriver
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service)
```

```python
URL = "https://www.amazon.com/s?k=laptops&__mk_es_US=%C3%85M%C3%85%C5%BD%C3%95%C3%91&crid=1PT1MO3SZ1SE7&sprefix=laptop%2Caps%2C313&ref=nb_sb_noss_1"
```

```python
driver.get(URL)
```

```python
def get_html(url, selector, sleep_time=20):
    driver.get(url)
    time.sleep(sleep_time)
    element = driver.find_element(By.CSS_SELECTOR, selector) 
    return element.get_attribute("outerHTML")
```

```python
html_content = get_html(URL, "#search > div.s-desktop-width-max.s-desktop-content.s-opposite-dir.s-wide-grid-style.sg-row > div.sg-col-20-of-24.s-matching-dir.sg-col-16-of-20.sg-col.sg-col-8-of-12.sg-col-12-of-16 > div > span.rush-component.s-latency-cf-section > div.s-main-slot.s-result-list.s-search-results.sg-row")
```

```python
soup = BeautifulSoup(html_content, "html.parser")
```

```python
h2_tags = soup.find_all("h2", attrs={"aria-label": True})
```

```python
for tag in h2_tags:
    print(tag["aria-label"])
```

```python
for a in soup.find_all("a"):
    if a.find("span", class_="a-offscreen"):
        print(a.find("span", class_="a-offscreen").get_text())
```

```python
url_twitter = "https://x.com/DanielNoboaOk"
```

```python
driver.get(url_twitter)
```

```python
cards = driver.find_elements(By.XPATH, '//article[@data-testid="tweet"]')
```

```python
cards
```

```python
tweets = []

for card in cards:
    texto = card.find_element(By.XPATH, './/div[@data-testid="tweetText"]').text
    fecha = card.find_element(By.TAG_NAME, 'time').get_attribute('datetime')
    tweets.append({
        'texto': texto,
        'fecha': fecha
    })
```

```python
df = pd.DataFrame(tweets)
```
