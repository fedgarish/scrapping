## WEB SCRAPPING

```python
import requests
from bs4 import BeautifulSoup
import sqlite3
```

```python
# Paso 1: Conectamos a la base de datos (o la creamos)
conn = sqlite3.connect('books.db')
cursor = conn.cursor()

# Paso 2: Creamos una tabla
cursor.execute('''
CREATE TABLE IF NOT EXISTS books (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    price TEXT,
    availability TEXT
)
''')
```

```python
url = 'https://books.toscrape.com/index.html'
```

```python
response = requests.get(url)
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.select('article.product_pod')
```

```python
books = soup.select('article.product_pod')
```

```python
books[0].h3.a['title']
```

```python
books[0].select_one('.price_color').text
```

```python
books[0].select_one('.availability').text.strip()
```

```python
for book in books:
    title = book.h3.a['title']
    # print(title)
    price = book.select_one('.price_color').text
    availability = book.select_one('.availability').text.strip()

    # Paso 4: Insertamos los datos en la base
    cursor.execute('INSERT INTO books (title, price, availability) VALUES (?, ?, ?)',
                   (title, price, availability))
```

```python
# Guardamos y cerramos
conn.commit()
conn.close()
print("¡Datos guardados en la base de datos!")
```
